```python
npl_data['Powerplay1_runrate'] = npl_data['Powerplay1_runs']/6
npl_data['Powerplay2_runrate'] = npl_data['Powerplay2_runs']/6
```

```python
grouped1 = npl_data.groupby('Team 1')['Powerplay1_runrate'].mean()
grouped1
```

```python
grouped2 = npl_data.groupby('Team 2')['Powerplay2_runrate'].mean()
grouped2
```

```python
group1_df = grouped1.reset_index()

group2_df = grouped2.reset_index()

merged_df = group1_df.merge(group2_df, left_on='Team 1', right_on='Team 2', how='inner')

merged_df['Average_Run_Rate_of_Powerplay'] = (merged_df['Powerplay1_runrate'] + merged_df['Powerplay2_runrate']) / 2

final_group_runrate_powerplay = merged_df[['Team 1', 'Average_Run_Rate_of_Powerplay']].rename(columns={'Team 1': 'Team'})
print(final_group_runrate_powerplay)
```

```python
plotbar(final_group_runrate_powerplay['Team'], final_group_runrate_powerplay['Average_Run_Rate_of_Powerplay'], 'Teams', 'Runrates by Team', 'Average Run rate of powerplay')
```

```python
grouped1 = npl_data.groupby('Team 1')['Run_Rate_Team_1'].mean()
grouped2 = npl_data.groupby('Team 2')['Run_Rate_Team_2'].mean()
group1_df = grouped1.reset_index()

group2_df = grouped2.reset_index()

merged_df = group1_df.merge(group2_df, left_on='Team 1', right_on='Team 2', how='inner')

merged_df['Average_Run_Rate'] = (merged_df['Run_Rate_Team_1'] + merged_df['Run_Rate_Team_2']) / 2

final_group_runrate = merged_df[['Team 1', 'Average_Run_Rate']].rename(columns={'Team 1': 'Team'})

print(final_group_runrate)
```

```python
plotbar(final_group_runrate['Team'], final_group_runrate['Average_Run_Rate'], 'Teams', 'Runrates by Team', 'Average Run rate')
```

```python
fig, axs = plt.subplots(34, 1, figsize=(12, 100))  
overs = [0,6, 20]

for i in range(min(len(npl_data['Team 1']), 32)):  
    powerplay1_runs = npl_data.iloc[i]['Powerplay1_runs']
    powerplay2_runs = npl_data.iloc[i]['Powerplay2_runs']
    target = npl_data.iloc[i]['Target_Runs']
    runs_scored_by_second_team = npl_data.iloc[i]['Run_Scored_By_2nd_Team']

    team1 = npl_data.iloc[i]['Team 1']
    team2 = npl_data.iloc[i]['Team 2']
    # Plotting with dots
    axs[i].plot(overs, [0,powerplay1_runs, target], marker='o', label=f'{team1}')  
    axs[i].plot(overs, [0,powerplay2_runs, runs_scored_by_second_team], marker='o', label=f'{team2}') 
    
    # Annotating the scores
    axs[i].annotate(f"{powerplay1_runs}", (overs[1], powerplay1_runs), textcoords="offset points", xytext=(-10, 5), ha='center')
    axs[i].annotate(f"{target}", (overs[2], target), textcoords="offset points", xytext=(-10, 5), ha='center')
    axs[i].annotate(f"{powerplay2_runs}", (overs[1], powerplay2_runs), textcoords="offset points", xytext=(-10, 5), ha='center')
    axs[i].annotate(f"{runs_scored_by_second_team}", (overs[2], runs_scored_by_second_team), textcoords="offset points", xytext=(-10, 5), ha='center')
    
    axs[i].set_title(f"Match {i + 1}")  
    axs[i].legend()  
    axs[i].set_xlabel("Overs")
    axs[i].set_ylabel("Runs")

plt.tight_layout()  
plt.show()
```

```python
def plotbar(teams, values,xlabel, ylabel, title):
   plt.bar(teams, values, color ='blue')
   plt.xlabel(xlabel)
   plt.ylabel(ylabel)
   plt.title(title)
   plt.xticks(rotation = 90)

   for i , value in enumerate(values):
       plt.text(i, value+0.1, f'{value:.2f}', ha = 'center', va ='bottom', fontsize = 10)
   plt.show()
```
